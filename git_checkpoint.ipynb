```python
# Git checkpoint for homepage implementation
This notebook runs git commands to check status, commit changes, and push to `origin develop`.
</VSCode.Cell>
<VSCode.Cell language="python">
import subprocess, os, textwrap
root = r"d:\Projects\usman-farooqi-portfolio"
def run(cmd):
    print(f"$ {cmd}")
    out = subprocess.run(cmd, cwd=root, shell=True, capture_output=True, text=True)
    print(out.stdout + out.stderr)
    return out

# 1. Run git status
run("git status --short --branch")
</VSCode.Cell>
<VSCode.Cell language="python">
# 2. Show current branch
run("git branch --show-current")
</VSCode.Cell>
<VSCode.Cell language="python">
# 3. Stage all changes
run("git add --all")
</VSCode.Cell>
<VSCode.Cell language="python">
# 4. Create commit
run("git commit -m \"Portfolio homepage v1 implementation\"")
</VSCode.Cell>
<VSCode.Cell language="python">
# 5. Show the new commit hash
run("git rev-parse HEAD")
</VSCode.Cell>
<VSCode.Cell language="python">
# 6. Push to origin develop
run("git push origin develop")
</VSCode.Cell>
<VSCode.Cell language="python">
# 7. Verify remote push success
run("git log --oneline origin/develop -1")
</VSCode.Cell>
<VSCode.Cell language="python">
# 8. Display exact terminal output
print("Notebook execution complete.")
```
